# Convolution 2D

```python
from scipy import signal
from scipy import misc
I = np.array([[8, 6, 2, 7], [6, 2, 4, 1], [5, 8, 5, 2], [3, 0, 3, 2]])
K = np.array([[4, 3], [7, 2]])
grad = signal.convolve2d(I, K, mode='valid')
```

# OpenCV Gamma Correction

```python
import cv2
import matplotlib.image as mpimg
import glob
```

```python
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
 
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)
```

```python
images = glob.glob('imgs//*.jpg')
```

```python
images
```

```python
image_plots = []
for i in images:
    image_plots.append(mpimg.imread(i))
```

```python
fig, gamma_correction_imgs = plt.subplots(1, 5, figsize=(10, 50))
for i in range(len(images)):
    gamma_correction_imgs[i].axis('off')
    gamma_correction_imgs[i].imshow(image_plots[i], cmap='gray')
```
